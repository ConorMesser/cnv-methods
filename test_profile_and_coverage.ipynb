### Create combined WES exome bin bed file

```python
bed_path = "../cnv_data/interval_files/exome_bins.phase1_2.20120518.consensus.hg38.bed"
bed_original = pd.read_csv(bed_path, sep='\t', header=None, names=['chr', 'start', 'end'])
```

```python
i = 0
bed_concat = []
while i < bed_original.shape[0]:
    curr_start = bed_original.iloc[i]['start']
    curr_end = bed_original.iloc[i]['end']
    while i+1 < bed_original.shape[0] and curr_end + 1 == bed_original.iloc[i+1]['start'] and bed_original.iloc[i]['chr'] == bed_original.iloc[i+1]['chr']:
        curr_end = bed_original.iloc[i+1]['end']
        i += 1
    bed_concat.append([bed_original.iloc[i]['chr'], curr_start, curr_end])
    i += 1
bed_concat = pd.DataFrame(bed_concat, columns=['chr', 'start', 'end'])
bed_concat.to_csv("../cnv_data/interval_files/exome_bins.phase1_2_combined.20120518.consensus.hg38.bed", sep='\t', index=False, header=False)
```

### Run WES coverage tools

```python
bed_path = "../cnv_data/interval_files/exome_bins.phase1_2_combined.20120518.consensus.hg38.bed"

tumor_bam_localization_task = wolf.localization.LocalizeToDisk(
  files = dict(
    bam = "../cnv_data/NA12878.alt_bwamem_GRCh38DH.20150826.CEU.exome.bam",
    bai = "../cnv_data/NA12878.alt_bwamem_GRCh38DH.20150826.CEU.exome.bam.bai"
  )
)

tumor_bam_localization = tumor_bam_localization_task.run()
```

```python
bed_df = pd.read_csv(bed_path, sep='\t', header=None, names=['chr', 'start', 'end'], dtype={0: str, 1: int, 2: int})
bed_df['len'] = bed_df['end'] - bed_df['start']
scatter_num = 200
import numpy as np
num_partitions_chr  = np.ceil(bed_df.groupby('chr')['len'].sum() / bed_df['len'].sum() * scatter_num)
intervals_out = []
for c, i in num_partitions_chr.items():
    this_chr_bed = bed_df.loc[bed_df['chr'] == c].reset_index()
    approx_partition_len = this_chr_bed['len'].sum() / i
    
    start = this_chr_bed.iloc[0]['start']
    len_cum = 0
    for num, row in this_chr_bed.iterrows():
        len_cum += row['len']
        if num == this_chr_bed.shape[0] - 1:
            intervals_out.append([c, start, row['end'] + 10])
        elif len_cum > approx_partition_len:
            intervals_out.append([c, start, row['end'] + 10])
            start = this_chr_bed.iloc[num+1]['start']
            len_cum = 0
            

subset_intervals = pd.DataFrame(intervals_out, columns=['chr', 'start', 'end'])
```

```python
# get coverage
cov_collect = wolf.ImportTask(
  task_path = "git@github.com:getzlab/covcollect.git",
  task_name = "covcollect"
)

cov_collect_task = cov_collect.Covcollect(
  inputs = dict(
    bam = tumor_bam_localization["bam"],
    bai = tumor_bam_localization["bai"],
    intervals = bed_path,
    subset_chr = subset_intervals["chr"],
    subset_start = subset_intervals["start"],
    subset_end = subset_intervals["end"],
  ),
  retry=2,
  # resources={'mem':'2G'}
)

cov_collect_results = cov_collect_task.run()
```

```python
# gather coverage
cov_gather = wolf.Task(
  name = "gather_coverage",
  inputs = { "coverage_beds" : [cov_collect_results["coverage"]] },
  script = """cat $(cat ${coverage_beds}) > coverage_cat.bed""",
  outputs = { "coverage" : "coverage_cat.bed" },
  job_avoid = False,
  retry=2
)

cov_gather_results = cov_gather.run()
```
